```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
import datetime
```

```python
path    = sorted(glob('/ceaza/lucas/CROCO/DESALADORAS_RUND0/CROCO_FILES/*blk_Y2001*'))
order   = [0,4,5,6,7,8,9,10,11,1,2,3]
path    = [path[n] for n in order]
blkera5 = [xr.open_dataset(p) for p in path]
```

```python
fig,ax = plt.subplots(1,len(blkera5[3].keys()), figsize=(30,4))
colors=plt.cm.tab10(np.linspace(0,1,len(blkera5[3].keys())))
for i,v in enumerate(blkera5[3].keys()):
    blkera5[3].isel(bulk_time=5)[v].plot.hist(ax=ax[i], density=True, bins=20, color=colors[i], edgecolor='k')
```
